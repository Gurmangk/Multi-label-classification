```python
median = df['Number of sexual partners'].median()
df['Number of sexual partners'].fillna(median, inplace=True)

mean = df['First sexual intercourse'].mean()
df['First sexual intercourse'].fillna(mean, inplace=True)

mean = df['Num of pregnancies'].mean()
df['Num of pregnancies'].fillna(mean, inplace=True)

median = df['Smokes (packs/year)'].median()
df['Smokes (packs/year)'].fillna(median, inplace=True)

median = df['Hormonal Contraceptives (years)'].median()
df['Hormonal Contraceptives (years)'].fillna(median, inplace=True)

mean = df['IUD (years)'].mean()
df['IUD (years)'].fillna(mean, inplace=True)

median = df['STDs (number)'].median()
df['STDs (number)'].fillna(median, inplace=True)

median = df['STDs:condylomatosis'].median()
df['STDs:condylomatosis'].fillna(median, inplace=True)

median = df['STDs:cervical condylomatosis'].median()
df['STDs:cervical condylomatosis'].fillna(median, inplace=True)

median = df['STDs:vaginal condylomatosis'].median()
df['STDs:vaginal condylomatosis'].fillna(median, inplace=True)

median = df['STDs:vulvo-perineal condylomatosis'].median()
df['STDs:vulvo-perineal condylomatosis'].fillna(median, inplace=True)

median = df['STDs:syphilis'].median()
df['STDs:syphilis'].fillna(median, inplace=True)

median = df['STDs:pelvic inflammatory disease'].median()
df['STDs:pelvic inflammatory disease'].fillna(median, inplace=True)

median = df['STDs:genital herpes'].median()
df['STDs:genital herpes'].fillna(median, inplace=True)

median = df['STDs:molluscum contagiosum'].median()
df['STDs:molluscum contagiosum'].fillna(median, inplace=True)

median = df['STDs:AIDS'].median()
df['STDs:AIDS'].fillna(median, inplace=True)

median = df['STDs:HIV'].median()
df['STDs:HIV'].fillna(median, inplace=True)

median = df['STDs:Hepatitis B'].median()
df['STDs:Hepatitis B'].fillna(median, inplace=True)

median = df['STDs:HPV'].median()
df['STDs:HPV'].fillna(median, inplace=True)

median = df['STDs: Number of diagnosis'].median()
df['STDs: Number of diagnosis'].fillna(median, inplace=True)

median = df['STDs: Time since first diagnosis'].median()
df['STDs: Time since first diagnosis'].fillna(median, inplace=True)

median = df['STDs: Time since last diagnosis'].median()
df['STDs: Time since last diagnosis'].fillna(median, inplace=True)

median = df['Dx:Cancer'].median()
df['Dx:Cancer'].fillna(median, inplace=True)

median = df['Dx:CIN'].median()
df['Dx:CIN'].fillna(median, inplace=True)

median = df['Dx:HPV'].median()
df['Dx:HPV'].fillna(median, inplace=True)

median = df['Dx'].median()
df['Dx'].fillna(median, inplace=True)
```
